```python
import pandas as pd

json_file = 'output.json'
json_df = pd.read_json(json_file)

# Path to CSV file
csv_file = 'data.csv'

# Save DataFrame to CSV file
json_df.to_csv(csv_file, index=False)
df= pd.read_csv(csv_file)
df = df[df['revenue']<=100_000_000]
df['revenue']
```

```python
# Filtering rows where 'industries' column does not contain "Artificial"
unique_industries = df[~df['industries'].str.contains("Artificial", case=False)]
unique_industries.reset_index(drop=True, inplace=True)
unique_industries


```

```python
unique_industries_subset = unique_industries[['name','emailAddresses','industries']].dropna().drop_duplicates()

#reset industries
unique_industries_subset.reset_index(drop=True, inplace=True)
print(unique_industries_subset)
```

```python
unique_industries_subset['industries'].explode().unique()
```

```python

# Filter rows where "Web Design Companies" is present in the "industries" column
web_design_companies_info = unique_industries_subset[unique_industries_subset['industries'].str.contains('Web Design Companies')].reset_index(drop=True)
print(web_design_companies_info[['name','emailAddresses']])

```

```python
contact_strings = []
for entry in web_design_companies_info['emailAddresses']:
    # Assuming each entry is a string representation of a dictionary
    # Split the string by commas to separate key-value pairs
    key_value_pairs = entry.strip('{}').split(', ')
    contact_string = None
    for pair in key_value_pairs:
        key, value = pair.split(': ')
        if key.strip("'") == 'contactString':
            contact_string = value.strip("'")
            break  
    
    contact_strings.append(contact_string)

print(contact_strings)
```

```python

web_design_companies_info['EmailAddress'] = contact_strings
print(web_design_companies_info[['name','EmailAddress']])
```

```python
web_design_companies_info[['name', 'EmailAddress',]].to_csv('web_design_companies_info.csv', index=False)

```
